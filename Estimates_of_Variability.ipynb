# **Deviations(độ lệch ) : The diffirent between the observed values and the estimate of location**

Những sai lêch của observed data với estimate of location (mean or median) cho chúng biết dữ liệu phân tán xung quanh giá trị trung tâm như thế nào 

Mean absolution deviation: Screenshot 2022-12-08 142828.png
Độ lệch tuyệt đối 
/n
Ko lấy trung bình sai số vì tích cực sẽ bù trừ cho tiêu cực -> trung bình trị tuyệt đối

```python
def MAD(arr):
  sum = 0
  mean = np.mean(arr)
  for i in arr:
    sum+= np.abs(i - mean)
  return sum/len(arr)
```

```python
MAD(data['Population'])
```

```python
sample = [12, 42, 53, 13, 112]
```

```python
MAD(sample)
```

Variance ( Phương sai ) Screenshot 2022-12-08 150010.png

```python
def Var(arr):
  sum = 0
  mean = np.mean(arr)
  for i in arr:
    sum+= (i-mean)*(i-mean)
  return sum/(len(arr)-1)
```

```python
Var(sample)
```
